```python
from bs4 import BeautifulSoup  
import pandas as pd
import requests
import re

keyword = input('지식인 검색어를 입력 해 주세요: ')
max_page = int(input('몇 페이지 까지 크롤링 할까요?: ')) # 5페이지
main_url = 'https://kin.naver.com/search/list.nhn?query='
page_url = '&page='

title_lst=[]
content_lst=[]
answer_lst=[]
# 1. 검색 결과 n페이지를 봐야하고 (검색어가 있어야 한다.)
for page_num in range(1,(max_page+1)):
    html = requests.get(main_url+keyword+page_url+str(page_num))
    soup = BeautifulSoup(html.text,'html.parser')
    # 하이퍼링크 추출
    ul_tag = soup.find('ul',class_='basic1')
    url_tag = ul_tag.find_all("dt")
    sub_url_lst = []
    for suburl in url_tag:
        sub_url_lst.append(suburl.find('a')['href'])
    
    #2. n 하이퍼링크에 연결된 게시물로 이동
    for nowurl in sub_url_lst:
        html_sub = requests.get(nowurl)
        soup_sub = BeautifulSoup(html_sub.text,'html.parser')
        # 2-1. 제목을 추출
        title = soup_sub.find('div',class_='title').text
        title = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣 ]',"",title)
        # 2-2. 본문을 추출(만약 본문이 없으면 '본문 없음')
        # print(nowurl)
        try: 
            content = soup_sub.find('div',class_='c-heading__content').text
            content = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣 ]',"",content)
        except AttributeError:
            content = ''
        # 2-3. 답변을 추출
        answer_text_lst = soup_sub.find_all("div",class_='_endContents c-heading-answer__content')
        # 2-3-1. 여러개의 답변일 경우 하나의 문자열로 만든다
        answer_text = ""
        for i in answer_text_lst:
            answer = re.sub('[^a-zA-Zㄱ-ㅣ가-힣 ]','',i.text)
            answer_text = answer_text + " " + answer
        
        # 3. 추출한 제목,본문,답변을 리스트에 추가
        title_lst.append(title)
        content_lst.append(content)
        answer_lst.append(answer_text)

Kin_df = pd.DataFrame({'제목':title_lst, '본문':content_lst,'답변':answer_lst})
Kin_df.to_csv('c:/pydata/송기라'+keyword+"지식인 크롤링.csv",encoding='cp949')
```
